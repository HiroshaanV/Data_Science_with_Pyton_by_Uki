```python
import csv #install csv package
import pandas as pd #install pandas package
result = pd.read_csv('result.csv') # read the csv file

ass1List=[] # define the list
ass2List=[] # define the list
ass3List=[] # define the list
ass4List=[] # define the list
examList=[] # define the list
totalList=[] # define the list
with open('result.csv') as csv_file: # open the csv file
    csv_reader = csv.reader(csv_file, delimiter=',') # read the csv file
    first = True
    index=0
    for row in csv_reader:
        if first:
            first=False
        else:
            ass1List.insert(index, row[1]) # assign the value to ass1 list
            ass2List.insert(index, row[2]) # assign the value to ass2 list
            ass3List.insert(index, row[3]) # assign the value to ass3 list
            ass4List.insert(index, row[4]) # assign the value to ass4 list
            examList.insert(index, row[5]) # assign the value to exam list
            totalList.insert(index, row[6]) # assign the value to total list
            index=index+1

def calculateAverageMinMax(listArray):  # define the function for claculate average, minimum & maximum
    sum=float(listArray[0])
    minimum=float(listArray[0])
    maximum=float(listArray[0])
    for i in range(1,len(listArray)):
        val=float(listArray[i])
        sum=sum+val
        if val<minimum:
            minimum=val
        if val>maximum:
            maximum=val
    average=sum/len(listArray)
    return [average,minimum,maximum]

ass1Answer = calculateAverageMinMax(ass1List) #assign the value
ass2Answer = calculateAverageMinMax(ass2List) #assign the value
ass3Answer = calculateAverageMinMax(ass3List) #assign the value
ass4Answer = calculateAverageMinMax(ass4List) #assign the value
examAnswer = calculateAverageMinMax(examList) #assign the value
totalAnswer = calculateAverageMinMax(totalList) #assign the value

print("Ass1 average: ",ass1Answer[0])
print("Ass1 min: ",ass1Answer[1])
print("Ass1 max: ",ass1Answer[2])
print("\n")
print("Ass2 average: ",ass2Answer[0])
print("Ass2 min: ",ass2Answer[1])
print("Ass2 max: ",ass2Answer[2])
print("\n")
print("Ass3 average: ",ass3Answer[0])
print("Ass3 min: ",ass3Answer[1])
print("Ass3 max: ",ass3Answer[2])
print("\n")
print("Ass4 average: ",ass4Answer[0])
print("Ass4 min: ",ass4Answer[1])
print("Ass4 max: ",ass4Answer[2])
print("\n")
print("Exam average: ",examAnswer[0])
print("Exam min: ",examAnswer[1])
print("Exam max: ",examAnswer[2])
print("\n")
print("Total average: ",totalAnswer[0])
print("Total min: ",totalAnswer[1])
print("Total max: ",totalAnswer[2])
print("\n")

df = pd.DataFrame(result) # define the dataframe
print("student with higest total: ", "\n",df[df.Total == df.Total.max()]) # print the maximum marks in the total column
```
