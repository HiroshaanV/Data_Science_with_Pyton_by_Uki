```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import scipy.stats
import pandas as pd

%matplotlib inline

df = pd.DataFrame({'a': np.random.randint(0, 50, size=100)}) # initialize the dataframe
df['b'] = df['a'] + np.random.normal(0, 10, size=100)
df['c'] = 100 - (2* df['a'] + np.random.normal(0, 10, size=100))
df['d'] = np.random.randint(0, 50, 100)

fig, ax = plt.subplots(figsize=(7, 5), dpi=100) # plotting
ax.scatter(df['a'],df['b'], alpha=0.6, edgecolor='none', s=100) # scatter plot
ax.set_xlabel('a samples') # name x axis
ax.set_ylabel('b samples') # name y axis

line_coef = np.polyfit(df['a'], df['b'], 1) # draw stright line
xx = np.arange(1, 50, 0.1)
yy = line_coef[0]*xx + line_coef[1]
ax.plot(xx, yy, 'r', lw=2)

fig, ax = plt.subplots(figsize=(7, 5), dpi=100) # plotting
ax.scatter(df['a'],df['c'], alpha=0.6, edgecolor='none', s=100) # scatter plot
ax.set_xlabel('a samples') # name x axis
ax.set_ylabel('c samples') # name y axis

line_coef = np.polyfit(df['a'], df['c'], 1) # draw stright line
xx = np.arange(1, 50, 0.1)
yy = line_coef[0]*xx + line_coef[1]
ax.plot(xx, yy, 'g', lw=2)

fig, ax = plt.subplots(figsize=(7, 5), dpi=100) # plotting
ax.scatter(df['a'],df['d'], alpha=0.6, edgecolor='none', s=100) # scatter plot
ax.set_xlabel('a samples') # name x axis
ax.set_ylabel('d samples') # name y axis

line_coef = np.polyfit(df['a'], df['d'], 1) # draw stright line
xx = np.arange(1, 50, 0.1)
yy = line_coef[0]*xx + line_coef[1]
ax.plot(xx, yy, 'y', lw=2)

pearson_r = np.cov(df['a'], df['b'])[0, 1] / (df['a'].std() * df['b'].std()) # calculatting pearson correlation
print("a and b pearson_r:", pearson_r)
print("a and b corrcoef:",np.corrcoef(df['a'],df['b'])) # calculating correlation coefficient

pearson_r = np.cov(df['a'], df['c'])[0, 1] / (df['a'].std() * df['c'].std()) # calculatting pearson correlation
print('\n',"a and c pearson_r:", pearson_r) 
print("a and c corrcoef:",np.corrcoef(df['a'],df['c'])) # calculating correlation coefficient

pearson_r = np.cov(df['a'], df['d'])[0, 1] / (df['a'].std() * df['d'].std()) # calculatting pearson correlation
print('\n',"a and d pearson_r:", pearson_r)
print("a and d corrcoef:",np.corrcoef(df['a'],df['d'])) # calculating correlation coefficient
```
