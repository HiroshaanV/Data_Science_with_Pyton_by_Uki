```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix

col_headers = []
for i in range(58):
  col_headers.append(i)

#Decision Trees 
spambase = pd.read_csv('spambase.data', header=None, names=col_headers)

spambase = spambase.rename(columns={57:'Target'})
target_col = spambase['Target']
features = spambase.drop('Target', axis=1)
x_train, x_test, y_train, y_test = train_test_split(features, target_col, random_state=0, test_size=0.2)

clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
print("Accuracy: ", clf.score(x_test, y_test),"percentage:",round((clf.score(x_test, y_test)*100), 2), "%" )
cm = plot_confusion_matrix(clf, x_test, y_test)
```

```python
#Using randomForest classifier
rf = RandomForestClassifier(n_estimators=10)
rf.fit(x_train, y_train)
rf.score(x_test, y_test)
print("Accuracy: ", rf.score(x_test, y_test), "percentage", round((rf.score(x_test, y_test)*100), 2), "%")

import matplotlib.pyplot as plt
%matplotlib inline

estimators = []
scores = []
for i in range(1, 21):
  rf = RandomForestClassifier(n_estimators=i)
  rf.fit(x_train, y_train)
  score = rf.score(x_test, y_test)
  estimators.append(i)
  scores.append(score)
plt.plot(estimators, scores);
```

```python
# Using support vector classification
svm_model = svm.SVC()
svm_model.fit(x_train, y_train)
svm_model.score(x_test, y_test)

svm_model = svm.SVC()
svm_model.fit(x_train, y_train)
print("Accuracy: ", svm_model.score(x_test, y_test), "(", round((svm_model.score(x_test, y_test)*100), 2), "%)\n")
cm = plot_confusion_matrix(svm_model, x_test, y_test)
```
