```python
import numpy as np
from matplotlib import pyplot as plt
labels = ['Cyber incident', 'Theft of paperwork or data storage device', 'Rogue employee','Social engineering / impersonation'] # assign the names to labels
colours = ['blue', 'red', 'green', 'yellow'] # assign the colors

total_attack_type = [108, 32, 26, 11] #assign the values for x axis
x_pos = np.arange(len(total_attack_type)) # assign the array for x axis position

fig,ax = plt.subplots(figsize=(7, 5)) #define the figure size
ax.bar(labels, total_attack_type, color=colours, align='center') #plot bar chart

ax.set_xlabel("Attack type") # label for x axis
ax.set_ylabel("Number of attacks per attack type") # label for y axis
ax.set_title("Number of malicious or criminal attack July-December-2019") # title for the plot

plt.xticks(x_pos, labels, rotation=90) # rotate 90 degrees of x axis labels
plt.show() 
```
